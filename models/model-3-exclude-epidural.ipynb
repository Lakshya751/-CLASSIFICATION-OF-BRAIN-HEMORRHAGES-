```python
import os
import numpy as np
import pandas as pd
import pydicom
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_curve
from tensorflow.keras import layers, models
from tensorflow.data import AUTOTUNE

###############################################################################
# 1. CONFIG & PATHS
###############################################################################
dataset_path = "/kaggle/input/rsna-intracranial-hemorrhage-detection/rsna-intracranial-hemorrhage-detection"
image_dir = os.path.join(dataset_path, "stage_2_train")
csv_path = os.path.join(dataset_path, "stage_2_train.csv")

SAMPLE_SIZE = 50000  
IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS = 15
LABEL_NAMES = ['intraparenchymal', 'intraventricular', 'subarachnoid', 'subdural']  # Removed epidural

###############################################################################
# 2. DATA PREPARATION
###############################################################################
df = pd.read_csv(csv_path)

# Convert "ID_000a0c3d5_subtype" format into columns
df['id'] = df['ID'].apply(lambda st: "ID_" + st.split('_')[1])
df['subtype'] = df['ID'].apply(lambda st: st.split('_')[2])
df = df[["id", "subtype", "Label"]].drop_duplicates()
df = df.pivot(index='id', columns='subtype', values='Label').reset_index()

for label in LABEL_NAMES:
    if label not in df.columns:
        df[label] = 0

df['file_path'] = df['id'].apply(lambda x: os.path.join(image_dir, f"{x}.dcm"))
df = df[df['file_path'].apply(os.path.exists)]

# Validate DICOM files
def validate_dicom_file(file_path):
    try:
        dcm = pydicom.dcmread(file_path)
        _ = dcm.pixel_array
        return True
    except Exception as e:
        print(f"Invalid DICOM file {file_path}: {str(e)}")
        return False

df['is_valid'] = df['file_path'].apply(validate_dicom_file)
df = df[df['is_valid']].drop(columns=['is_valid'])


if SAMPLE_SIZE is not None and SAMPLE_SIZE < len(df):
    df = df.sample(SAMPLE_SIZE, random_state=42).reset_index(drop=True)

print(f"Total slices in dataframe: {len(df)}")
print("Positives per subtype:")
print(df[LABEL_NAMES].sum())

# Stratified Train/Validation Split
df['stratify'] = df[LABEL_NAMES].apply(lambda x: ''.join(x.astype(str)), axis=1)
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['stratify'])
train_df = train_df.drop(columns=['stratify'])
val_df = val_df.drop(columns=['stratify'])

# Class-Specific Oversampling
target_positives = int(max(train_df[LABEL_NAMES].sum()))  
print(f"Target number of positives per class: {target_positives}")

balanced_train_dfs = []
for label in LABEL_NAMES:
    pos_samples = train_df[train_df[label] == 1]
    if len(pos_samples) == 0:
        print(f"No positive samples for {label}, skipping oversampling.")
        continue
    oversample_factor = max(1, target_positives // len(pos_samples))
    pos_oversampled = pos_samples.sample(n=oversample_factor * len(pos_samples), 
                                        replace=True, random_state=42)
    if len(pos_oversampled) > target_positives:
        pos_oversampled = pos_oversampled.sample(n=target_positives, random_state=42)
    balanced_train_dfs.append(pos_oversampled)

# Add negative samples
neg_train = train_df[train_df[LABEL_NAMES].sum(axis=1) == 0]
if len(neg_train) > 0:
    total_positives = target_positives * len(LABEL_NAMES)
    neg_sample = neg_train.sample(n=min(len(neg_train), total_positives), 
                                  replace=True, random_state=42)
    balanced_train_dfs.append(neg_sample)

# Combine and shuffle (no deduplication)
train_df = pd.concat(balanced_train_dfs).sample(frac=1, random_state=42)
print(f"After class-specific oversampling, training size: {len(train_df)}")
print("Training positives per subtype:")
print(train_df[LABEL_NAMES].sum())

###############################################################################
# 3. WINDOWING & NORMALIZATION
###############################################################################
def apply_window(image, center, width):
    lower_bound = center - width // 2
    upper_bound = center + width // 2
    windowed = tf.clip_by_value(image, lower_bound, upper_bound)
    imin = tf.reduce_min(windowed)
    imax = tf.reduce_max(windowed)
    return (windowed - imin) / (imax - imin + 1e-8)

###############################################################################
# 4. PARSE DICOM: MULTI-WINDOW EXAMPLE
###############################################################################
def parse_dicom_multiwindow(file_path):
    def _py_read_dicom(fp):
        path_str = fp.numpy().decode('utf-8')
        dcm = pydicom.dcmread(path_str)
        return dcm.pixel_array.astype(np.float32)
    
    image = tf.py_function(_py_read_dicom, [file_path], tf.float32)
    image.set_shape([None, None])
    
    brain = apply_window(image, center=40, width=80)
    subdural = apply_window(image, center=80, width=200)
    bone = apply_window(image, center=600, width=2800)
    
    multi = tf.stack([brain, subdural, bone], axis=-1)
    multi = tf.image.resize(multi, [IMG_SIZE, IMG_SIZE])
    return multi

###############################################################################
# 5. AUGMENTATION
###############################################################################
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.05),
    layers.RandomZoom(0.2),
    layers.RandomContrast(0.1),
])

###############################################################################
# 6. BUILD DATASET
###############################################################################
def create_dataset(dataframe, shuffle=False, augment=False, batch_size=32):
    file_paths = dataframe['file_path'].values
    labels = dataframe[LABEL_NAMES].values.astype(np.float32)
    
    ds_files = tf.data.Dataset.from_tensor_slices(file_paths)
    ds_labels = tf.data.Dataset.from_tensor_slices(labels)
    ds = tf.data.Dataset.zip((ds_files, ds_labels))
    
    def _load_and_augment(fp, lbl):
        img = parse_dicom_multiwindow(fp)
        if augment:
            img = data_augmentation(img)
        return img, lbl
    
    ds = ds.map(_load_and_augment, num_parallel_calls=AUTOTUNE)
    if shuffle:
        ds = ds.shuffle(buffer_size=1024)
    
    ds = ds.batch(batch_size).prefetch(AUTOTUNE)
    return ds

train_ds = create_dataset(train_df, shuffle=True, augment=True, batch_size=BATCH_SIZE)
val_ds = create_dataset(val_df, shuffle=False, augment=False, batch_size=BATCH_SIZE)

###############################################################################
# 7. FOCAL LOSS
###############################################################################
def focal_loss(gamma=2.0, alpha=None):
    if alpha is None:
        alpha = [0.5, 0.5, 0.75, 0.5]  
    def focal_loss_fixed(y_true, y_pred):
        y_pred = tf.clip_by_value(y_pred, 1e-7, 1.0 - 1e-7)
        ce = - (y_true * tf.math.log(y_pred) + (1 - y_true) * tf.math.log(1 - y_pred))
        weight = tf.zeros_like(y_true)
        for i, a in enumerate(alpha):
            weight += a * y_true[:, i:i+1] + (1 - a) * (1 - y_true[:, i:i+1])
        focal_factor = ((1 - y_pred) ** gamma) * y_true + (y_pred ** gamma) * (1 - y_true)
        return tf.reduce_mean(weight * focal_factor * ce)
    return focal_loss_fixed

###############################################################################
# 8. MODEL (EfficientNetB0)
###############################################################################
base_model = tf.keras.applications.EfficientNetB0(
    weights='imagenet',
    include_top=False,
    input_shape=(IMG_SIZE, IMG_SIZE, 3)
)
base_model.trainable = True

input_layer = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x = base_model(input_layer, training=True)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.3)(x)
output_layer = layers.Dense(len(LABEL_NAMES), activation='sigmoid')(x)

model = tf.keras.Model(inputs=input_layer, outputs=output_layer)
model.summary()

###############################################################################
# 9. TRAINING
###############################################################################
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-4),
    loss=focal_loss(gamma=2.0, alpha=[0.5, 0.5, 0.75, 0.5]),
    metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
)

callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True),
    tf.keras.callbacks.ModelCheckpoint("best_model.keras", save_best_only=True)
]

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=callbacks
)

###############################################################################
# 10. EVALUATION
###############################################################################
plt.figure(figsize=(18, 5))

# Accuracy
plt.subplot(1, 3, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title("Accuracy over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)

# Loss
plt.subplot(1, 3, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Loss over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)

# AUC
plt.subplot(1, 3, 3)
plt.plot(history.history['auc'], label='Train AUC')
plt.plot(history.history['val_auc'], label='Validation AUC')
plt.title("AUC over Epochs")
plt.xlabel("Epoch")
plt.ylabel("AUC")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

print("Final Training AUC:", history.history['auc'][-1])
print("Final Validation AUC:", history.history['val_auc'][-1])

val_images, val_labels_all = [], []
for batch_imgs, batch_labs in val_ds:
    val_images.append(batch_imgs)
    val_labels_all.append(batch_labs)

val_images = tf.concat(val_images, axis=0)
val_labels_all = tf.concat(val_labels_all, axis=0)

preds = model.predict(val_images)

# Threshold Tuning
thresholds = {}
for i, label in enumerate(LABEL_NAMES):
    fpr, tpr, thresh = roc_curve(val_labels_all[:, i], preds[:, i])
    optimal_idx = np.argmax(tpr - fpr)
    thresholds[label] = thresh[optimal_idx]
    print(f"Optimal threshold for {label}: {thresholds[label]:.3f}")

# Apply optimal thresholds
preds_binary = np.array([[1 if p >= thresholds[label] else 0 for p, label in zip(pred, LABEL_NAMES)] 
                         for pred in preds])

# Classification Report with Optimal Thresholds
print("\nClassification Report with Optimal Thresholds:")
print(classification_report(val_labels_all, preds_binary, target_names=LABEL_NAMES, zero_division=0))

# Standard Classification Report (Threshold 0.5)
preds_binary_05 = (preds > 0.5).astype(int)
true_labels = val_labels_all.numpy()
print("\nClassification Report (Threshold 0.5):")
print(classification_report(true_labels, preds_binary_05, target_names=LABEL_NAMES, zero_division=0))

accuracy = (preds_binary_05 == true_labels).mean()
print(f"\nAverage Binary Accuracy (Threshold 0.5): {accuracy * 100:.2f}%")
print("Done!")
```
